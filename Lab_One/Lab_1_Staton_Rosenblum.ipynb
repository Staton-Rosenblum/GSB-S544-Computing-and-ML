**Task One**

Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.



The aesthetics are income on the x-axis and life expectancy on the y-axis. Also, the color of the bubble represents the region of the world the data is from and the size of the bubble shows the population size of that given country for the corresponding data point.

Recreate the plot

```python
import pandas as pd
from plotnine import ggplot, aes, geom_point, scale_size, scale_color_manual, labs, theme_bw, scale_x_continuous, scale_y_continuous, annotate, theme, scale_fill_manual, scale_size

gray_diamond = '\u25C8'
df = pd.read_csv('q1data.csv')
df_clean = df.dropna()
region_colors = {
    "asia": "#FF5872",
    "europe": "#FFE700",
    "africa": "#00D5E9",
    "americas": "#7FEB00",
}
region_labels = {
    "asia": "Asia",
    "europe": "Europe",
    "americas": "Americas",
    "africa": "Africa"
}
plot = (
    ggplot(df_clean, aes(x="income", y="life_exp", fill="four_regions", size="population")) +
    geom_point(alpha=.9, stroke=.3, color = "black") +
    scale_x_continuous(
        trans='log2',
        breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],
        limits=(400, 64000),
        labels=['500', '1000', '2000', '4000', '8000', '16k', '32k', '64k'] #Used Chatgpt to find out how to set the x-axis increments
    ) +
    scale_y_continuous(
        breaks=range(20, 91, 10),
        limits=(20, 90) #Used Chatgpt to find out how to set the y-axis increments
    ) +
    scale_fill_manual(
        values=region_colors,
        name="World Regions",
        labels=region_labels
    ) +
        scale_size(range=(1, 25), name="Population" #USed CHATGPT to help make the size of the bubbles correspond to the population size
    ) +
    labs(
        x="Income",
        y="Life Expectancy",
        size="Bubble Size = Population",
        color="World Regions"
    ) +
    theme_bw()
    + annotate('text', x=6000, y=48, label='2010', size=200, alpha=0.2, color='gray')
    + annotate('text', x=40000, y=20, label= 'per person (GDP/capita, PPPS inflation-adjusted)', size=7, alpha=.9, color='gray')
    + annotate('text', x=400, y=90, label='years', size=7, color='gray', angle=90)
    + annotate('text', x=1300, y=90, label='INCOME LEVEL 1', size=9, alpha = .9, color='gray')
    + annotate('text', x=4700, y=90, label='LEVEL 2', size=9, alpha = .9, color='gray')
    + annotate('text', x=14000, y=90, label='LEVEL 3', size=9, alpha = .9, color='gray')
    + annotate('text', x=40000, y=90, label='LEVEL 4', size=9, alpha = .9, color='gray')
    + annotate('text', x=2500, y=90, label=gray_diamond, size=9, color='gray')
    + annotate('text', x=8000, y=90, label=gray_diamond, size=9, color='gray')
    + annotate('text', x=24000, y=90, label=gray_diamond, size=9, color='gray')
    + theme(aspect_ratio=0.5)
    + theme(figure_size=(10,6))
    + theme(legend_position='none')
)

plot




```

Below is another geometry we can use to visualize this data. Using a boxplot to visualize this data set is not as effective as the plot above. The first issue is that we can not see the distribution of the income variable for each region. The boxplots are centered around the mean, but we do not know the variability. The boxplot geometry does show us the distribution of the life expectancy variable, but it is not as easy to read as the bubble plot above. Lastly, we are not able to see each data point, so we are not able to see the population size of each individual country's data.

```python
import pandas as pd
from plotnine import ggplot, aes, geom_boxplot, scale_size, scale_color_manual, labs, theme_bw, scale_x_continuous, scale_y_continuous, annotate, theme, scale_fill_manual, scale_size

df = pd.read_csv('q1data.csv')
df_clean = df.dropna()
region_colors = {
    "asia": "#FF5872",
    "europe": "#FFE700",
    "africa": "#0EA5B4",
    "americas": "#7FEB00",
}
region_labels = {
    "asia": "Asia",
    "europe": "Europe",
    "americas": "Americas",
    "africa": "Africa"
}
plot = (
    ggplot(df_clean, aes(x="income", y="life_exp", fill="four_regions", size="population")) +
    geom_boxplot(alpha=.9) +
    scale_x_continuous(
        trans='log2',
        breaks=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000],
        limits=(500, 64000),
        labels=[500, 1000, 2000, 4000, 8000, 16000, 32000, 64000]
    ) +
    scale_y_continuous(
        breaks=range(20, 91, 10),
        limits=(20, 90)
    ) +
    scale_fill_manual(
        values=region_colors,
        name="World Regions",
        labels=region_labels
    ) +
    theme_bw()
    + annotate('text', x=6000, y=48, label='2010', size=200, alpha=0.2, color='gray')
    + annotate('text', x=40000, y=20, label= 'per person (GDP/capita, PPPS inflation-adjusted)', size=6, alpha=.9, color='gray')
    + theme(aspect_ratio=0.5)
    + theme(figure_size=(10,6))
    + theme(legend_position='none')
)

plot
```

**Task Two**

Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

We have exports as the x variable on the x-axis, imports on the y axis as the y variable, the size of the bubble corresponding to the energy use data, and the color of the bubbles represent the region the data is from.

```python
import pandas as pd
from plotnine import ggplot, aes, geom_point, scale_size, scale_color_manual, labs, theme_bw, scale_x_continuous, scale_y_continuous, annotate, theme, scale_fill_manual, scale_size

df2 = pd.read_csv('q2data.csv')
df_clean2 = df2.dropna()
region_colors = {
    "asia": "#FF5872",
    "europe": "#FFE700",
    "africa": "#00D5E9",
    "americas": "#7FEB00",
}
region_labels = {
    "asia": "Asia",
    "europe": "Europe",
    "americas": "Americas",
    "africa": "Africa"
}
def numeric_energy(value):
    if 'k' in value:
        return float(value.replace('k', '').strip()) * 1000
    else:
        return float(value) # Used CHATGPT to help clean the data to remove values like "10k" and convert it to 10000

df_clean2['energy'] = df_clean2['energy'].apply(numeric_energy)
plot = (
    ggplot(df_clean2, aes(x="exports", y="imports", fill="four_regions", size="energy")) +
    geom_point(alpha=.9, stroke=.3, color = "black") +
    scale_x_continuous(
        breaks=range(20, 239, 20),
        limits=(0,220),
    ) +
    scale_y_continuous(
        breaks=range(50, 401, 50),
        limits=(0,400) #Used Chatgpt to find out how to set the y-axis increments
    ) +
    scale_fill_manual(
        values=region_colors,
        name="World Regions",
        labels=region_labels
    ) +
        scale_size(range=(1, 25), name="energy"
    ) +
    labs(
        x="Exports (\% of GDP)",
        y="Imports (\% of GDP)",
        size="Bubble Size = Energy",
        color="World Regions"
    ) +
    theme_bw()
    + annotate('text', x=110, y=180, label='1997', size=200, alpha=0.2, color='gray')
    + theme(aspect_ratio=0.5)
    + theme(figure_size=(10,6))
    + theme(legend_position='none')
)

plot
```

**Task Three**

Identify, and list here, what aesthetics are being used and which variables are being mapped to each aesthetic.

The aesthetics are individuals using the internet as the x variable on the x-axis, GDP/capita on the y-axis as the y variable the size of the bubbles being the amount of income, and the color of the bubbles representing the region from the four_regions data.

```python
import pandas as pd
from plotnine import ggplot, aes, geom_point, scale_size, scale_fill_manual, labs, theme_bw, scale_x_continuous, scale_y_continuous, annotate, theme, scale_color_manual, scale_size

df3 = pd.read_csv('q3data.csv')
df_clean3 = df3.dropna()

region_colors = {
    "asia": "#FF5872",
    "europe": "#FFE700",
    "africa": "#0EA5B4",
    "americas": "#7FEB00",
}
region_labels = {
    "asia": "Asia",
    "europe": "Europe",
    "americas": "Americas",
    "africa": "Africa"
}
plot = (ggplot(df_clean3, aes(x='internet_users', y='gdp', fill='four_regions', size='income'))
    + geom_point()
    + scale_x_continuous(
        breaks=range(10, 91, 10),
        limits=(0, 100) )
    + scale_y_continuous(
        trans='log2',
        breaks=[200, 500, 1000, 2000, 5000, 10000, 20000, 50000, 100000],
        labels=("200", "500", "1000", "2000", "5000", "10k", "20k", "50k", "100k"),
        limits=(0, 200000))
    + scale_fill_manual(
        values=region_colors,
        name="World Regions",
        labels=region_labels) +
    scale_size(range=(1, 20), name="income"
    ) +
    labs (
        x="Individuals using the Internet",
        y="GDP/capita",
        size="Bubble Size = Average Household Income",
        color="World Regions"
      )
      + theme_bw()
      + annotate('text', x=50, y=5000, label='2001', size=200, alpha=0.2, color='gray')
      + annotate('text', x=1, y=95000, label='US$ inflation-adjusted', size=6, color='gray', angle=90)
      + annotate('text', x=95, y=200, label='\% of population', size=6, color='gray')
      + theme(aspect_ratio=0.5)
      + theme(figure_size=(10,6))
      + theme(legend_position='none')
      )
plot
```
